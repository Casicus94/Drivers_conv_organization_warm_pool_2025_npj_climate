```python
import xarray as xr 
import matplotlib.pyplot as plt
import numpy as np
import cartopy.io.shapereader as shpreader
import cartopy.crs as ccrs
from cartopy.feature import ShapelyFeature
import cartopy.feature as cfeature
import matplotlib.ticker as mticker
import matplotlib.gridspec as gridspec #Para generar paneles en las graficas!
from pylab import *
```

```python
ds = xr.open_dataset('../NN_real/TCWV_area.nc')
```

```python
fig = plt.figure(figsize=(10,8)) #ancho, largo
gs = gridspec.GridSpec(1,1)
ax = plt.subplot(gs[0], projection= ccrs.PlateCarree()) 
im = plt.contourf(ds.variables['longitude'],ds.variables['latitude'], ds.tcwv[647,:,:], 
                  transform = ccrs.PlateCarree(), levels = np.linspace(35,60,30), 
                  cmap = 'Spectral', extend = 'both') 
bar = plt.colorbar(im, shrink = 0.8)
bar.set_label('TCWV (mm)')
gl = ax.gridlines(crs = ccrs.PlateCarree(), draw_labels = True, linewidth = 1, 
                  color='white', alpha = 0.5, linestyle = '--')
gl.xlocator = mticker.FixedLocator([137.5, 140, 142.5]); gl.ylocator = mticker.FixedLocator([3, 4.5, 6, 7.5]); gl.ylabels_right = False;
```
